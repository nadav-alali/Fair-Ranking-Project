# Check type attributes

```python
type_atts = ['sex', 'age', 'race']
df_types = df[type_atts]
df_types['age_binary'] = df_types['age'].apply(lambda x: 0 if x < 35 else 1)
df_types
```

```python
df_types['race'].value_counts()
```

```python
attributes = scoring_attr
df = df.dropna(subset=attributes)

for col in attributes:
    min_val = df[col].min()
    max_val = df[col].max()
    df[col] = (df[col] - min_val) / (max_val - min_val)

# create a binary groups for 'age'
df['age_binary'] = df['age'].apply(lambda x: '<36' if x < 35 else '>=36')
df['age_bucketized'] = df['age'].apply(lambda x: '<30' if x < 30 else ('31-40' if 30 < x <= 40 else '>40'))
```

```python
df
```

```python
type_atts = ['sex', 'age_binary','age_bucketized', 'race']
df_types = df[type_atts]
df_types
```

```python
def balance_by_sex(df):
    # Split the DataFrame into two: one for Male and one for Female
    male_df = df[df['sex'] == 'Male']
    female_df = df[df['sex'] == 'Female']

    # Determine the minimum number of samples between the two groups
    min_count = min(len(male_df), len(female_df))

    # Sample min_count rows from each group (using a fixed random_state for reproducibility)
    balanced_male = male_df.sample(n=min_count, random_state=42)
    balanced_female = female_df.sample(n=min_count, random_state=42)

    # Combine the two groups and shuffle the result
    balanced_df = pd.concat([balanced_male, balanced_female]).sample(frac=1, random_state=42).reset_index(drop=True)
    return balanced_df
df_types = balance_by_sex(df_types)
counts_df = df_types.value_counts().reset_index(name='count')
# Compute the average of the counts
avg_count = counts_df['count'].sum()
# Add a new column where every row contains the average count
counts_df['count_avg'] = counts_df['count'] / avg_count
counts_df
```
